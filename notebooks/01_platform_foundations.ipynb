## 1. Verify schema and tables

### Startup check

This notebook verifies that the demo tables exist before running the platform queries. If the tables are missing, rerun `00_data_generation.ipynb` first.

```python
missing_tables = []
for table in ['campaigns', 'impressions', 'clicks', 'conversions', 'costs']:
    try:
        spark.sql(f"DESCRIBE TABLE adtech_demo.adtech_platform.{table}").show(1)
    except Exception:
        missing_tables.append(table)

if missing_tables:
    print('Missing demo tables:', missing_tables)
    print('Please rerun notebooks/00_data_generation.ipynb first to create and populate the tables.')
else:
    print('All demo tables exist. Ready to continue.')
```
